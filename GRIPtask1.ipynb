Predict the percentage of an student based on the no.of study hours.

```python
# I am going to import important lobraries for this work
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Now i am going to fatch my dataset 
url = "http://bit.ly/w-data"
ds_data = pd.read_csv(url)
print("importing is successfully")


ds_data.head(12)
```

```python
# Now we are try to understand the relation between data . I try to Plotting the distribution of score

ds_data.plot(x = 'Hours' , y = 'Scores' , style ='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage score')
plt.show()


```

Now we are going to Preparing the our dataset

```python
X = ds_data.iloc[:, :-1].values
Y = ds_data.iloc[:,1].values
```

Now we have split our dataset into training and testing data.


```python
# for split into training and testing data
# I am going to import Scikit Learn

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)

```

*NOW WE ARE READY TO TRAINING THE ALGORITHM*

```python
# now i am going to import LinearRegression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)


print("training complete.")
```

```python
# PLOTTING THE REGRESSION LINE
line = regressor.coef_*X+regressor.intercept_


# plotting for the test data 

plt.scatter(X,Y)
plt.plot(X,line);
plt.show()
```

MARKING PREDICTIONS

```python
print(X_test) #testing data - In hours
Y_pred = regressor.predict(X_test) # predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual':Y_test,'predicted': Y_pred})
df

```

```python
# You can also test with your own data
hours =np.array(2.4)
hours = hours.reshape(-1, 1)
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours[0]))
print("Predicted Score = {}".format(own_pred[0]))
```

Evaluating the model

```python
from sklearn import metrics
print('mean Absolute Error:',metrics.mean_absolute_error(Y_test,Y_pred))
```
